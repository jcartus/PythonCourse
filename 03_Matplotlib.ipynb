# Overview: Plotting and Matplotlib
- using matplotlib to plot a simple curve
- using matplotlib to do a quick histogram
- using matplotlib and seaborn to make surf plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Simple plot 

```python
x = np.arange(10)
y = x**2

plt.plot(x, y)
```

# Setting axis labels and a legend

```python
# generate the data
x = np.arange(10)
y1 = x**2
y2 = 2*x

# plot the curves. The label parameter is what is displyed in the legend
plt.plot(x, y1, label="y1")
plt.plot(x, y2, label="y2")


# set the axis labels
plt.xlabel("x / a.u.")
plt.ylabel("f(x) / a.u.")

# this adds a legend to the plot.
# the contents of the legend are defined via the "label"-parameters
# in the plot command
plt.legend()
```

# Setting limits 

```python
x = np.arange(10)
y = x**2

plt.plot(x, y)

plt.xlim(4, 6)
```

# Custom styles
You can set the line style (dashed, points of straight line), color and marker individually. This can be done as parameters in the plot function.

- colors: https://matplotlib.org/3.1.0/gallery/color/named_colors.html
- Linestyles: https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html
- Markers: https://matplotlib.org/3.1.1/api/markers_api.html

```python
x = np.arange(10)
y = x**2

plt.plot(x, y, color="green", marker="x", linestyle="dashed")
```

There is a shorthand notation where a string, containing this information, is passed to the plot function. 

```python
x = np.arange(10)
y = x**2


# g is for the color green
# x is for the x mater
# -- is for the dashed line
plt.plot(x, y, "gx--")
```

# Subplots 
It can be helpful to plot a few things together at once. this can be done wiht subplots(n, m, i). n, m specifies the number shape of a matrix of plots; i specifies the index.

The below examples creates a plot of two plots on top of each other

```python
#--- define the first plot ---
plt.subplot(2, 1, 1)

plt.plot(x, x**2)
plt.xlabel("x")
plt.ylabel("y")
#---

#--- define the second plot ---
plt.subplot(2, 1, 2)

plt.plot(x, x**2)
plt.xlabel("x")
plt.ylabel("y")
#---
```

# surface plot 
You will need 3 arrays one for x and for y and one for z. 

to get the x and y combinations numpy's meshgrid is very usefull

```python
x = np.arange(5)
y = np.linspace(0, 2, 8)
```

```python
x
```

```python
y
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
# xx and yy contian all combination of x and y in suitable order. 
# therefore they are of equal shape.
xx.shape, yy.shape
```

```python
xx
```

```python
yy
```

```python
zz = xx**2 * yy
```

```python
# simple contours
plt.contour(xx, yy, zz)
```

```python
# filled contours
plt.contourf(xx, yy, zz)
```
